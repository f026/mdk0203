# Задание 1

Предприятие, в состав которого входит ряд подразделений, имеет ремонтную мастерскую, в которой производится различного вида ремонт средств вычислительной и оргтехники. Необходимо вести учет, сданной в ремонт техники. Для каждой единицы техники необходимо хранить её инвентарный номер, название устройства, его модель, год выпуска, какому подразделению устройство принадлежит на текущую дату, т.е. необходимо хранить историю перемещения техники между подразделениями организации. Перемещение единицы техники осуществляется на основе документа, в котором указано с какой даты данная единица перемещается в указанное подразделение. Эта дата одновременно является датой окончания нахождения единицы техники в предыдущем подразделении. Необходимо также хранить дату сдачи единицы техники в ремонт, вид ремонта, срок ремонта, ФИО, номер сотрудника, сдавшего технику в ремонт, ФИО, номер сотрудника, который принял технику в ремонт, ФИО, номер, должность сотрудника, выполняющего ремонт, перечень необходимых для ремонта запасных частей, их стоимость на текущую дату. Стоимость запасных частей определяется из документа – накладной, по которой они получены, в ней и указана дата получения, стоимость. О каждом сотруднике необходимо хранить информацию о всех перемещениях за время работы на данном предприятии: дата начала работы, дата окончания, подразделение, должность.

Построить диаграмму, которая показывает количество отремонтированной техники по каждому подразделению

```python
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib.patches as mpatches

data = {
    'Подразделение': ['Офис', 'Технический отдел', 'Производство', 'Бухгалтерия'],
    'Количество ремонта': [15, 20, 10, 8]
}

df = pd.DataFrame(data)


# Создаем цвета для каждой категории
colors = ['blue', 'green', 'red', 'purple']

# Создаем прямоугольники с разными цветами
rects = plt.bar(df['Подразделение'], df['Количество ремонта'], color=colors)

# Добавляем подписи к столбцам
for rect in rects:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2., height,
             f'{height}',
             ha='center', va='bottom')

# Создаем легенду
legend_patches = [mpatches.Patch(color=color, label=f"{df.loc[i, 'Подразделение']}")
                  for i, color in enumerate(colors)]

plt.legend(handles=legend_patches)

plt.title('Количество отремонтированной техники по подразделениям')
plt.xlabel('Подразделение')
plt.ylabel('Количество ремонта')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

Чтобы облегчить труд работников библиотеки была разработана информационная система, которая должна обеспечивать возможность создание новых бланков, вести точный учет читателей, выданных и возвращенных книг, выполненных работ, корректировать данные по читателям, осуществлять быстрый поиск информации.

Построить диаграмму, которая показывает сколько книг взял каждый читатель


```python
import matplotlib.pyplot as plt

# Создание данных
readers = ['Иван', 'Мария', 'Петр', 'Анна', 'Сергей']
books = ['Книга 1', 'Книга 2', 'Книга 3', 'Книга 4', 'Книга 5']

borrowed_books = {
    'Иван': ['Книга 1', 'Книга 2'],
    'Мария': ['Книга 3', 'Книга 4'],
    'Петр': ['Книга 5'],
    'Анна': [],
    'Сергей': ['Книга 1']
}

# Функция для подсчета взятых книг
def count_borrowed_books(reader):
    return len(borrowed_books.get(reader, []))

# Создание списка значений для диаграммы
values = [count_borrowed_books(reader) for reader in readers]

# Создание столбчатой диаграммы
plt.figure(figsize=(10, 6))
plt.bar(readers, values)
plt.xlabel('Читатели')
plt.ylabel('Количество взятых книг')
plt.title('Количество взятых книг у каждого читателя')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Задание 4

Компания – торговый посредник, продающая товары различных производителей, разрабатывает систему обработки заказов. Дважды в год компания публикует каталог продуктов, который рассылается клиентам и другим заинтересованным лицам.
Клиенты приобретают товары, направляя в компанию перечень продуктов с информацией об оплате. Компания выполняет заказы и отправляет товары по адресам клиентов.
Система должна отслеживать заказ от момента его получения до отправки товара. Клиенты могут возвращать товары, оплачивая, возможно, при этом некоторые издержки. Часть клиентов заказывает товары через Интернет. Компания пользуется услугами различных транспортных и страховых компаний.




```python
import matplotlib.pyplot as plt
import numpy as np

# Фиктивные данные
# - продажи за последние 6 месяцев в тысячах рублей
sales = [1000, 1200, 1500, 1800, 2000, 2200]
months = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь']

# - количество заказов в месяц
orders = [500, 550, 600, 650, 700, 750]

# - цены на товары в рублях
prices = [10, 15, 20, 25, 30, 35]

# - время доставки (дни)
delivery_times = [5, 7, 3, 4, 6, 8]

# - процент возвратов товара
returns_percent = [2, 2.5, 3, 3.5, 4, 4.5]

# График продаж
plt.figure(figsize=(10, 6))
plt.plot(months, sales, marker='o')
plt.title('Среднемесячные продажи (тыс. руб.)')
plt.xlabel('Месяц')
plt.ylabel('Сумма продаж (тыс. руб.)')
plt.grid(True)
plt.show()

# Диаграмма разброса заказов
plt.figure(figsize=(10, 6))
plt.scatter(range(len(orders)), orders, s=50, alpha=0.7)
plt.title('Количество заказов в месяц')
plt.xlabel('Месяц')
plt.ylabel('Количество заказов')
plt.grid(True)
plt.show()

# Столбчатая диаграмма цен на товары
plt.figure(figsize=(10, 6))
plt.bar(months, prices, color='skyblue', edgecolor='black')
plt.title('Цены на товары (руб.)')
plt.xlabel('Месяц')
plt.ylabel('Цена (руб.)')
plt.grid(axis='y', linestyle='--')
plt.show()

# Гистограмма времени доставки
plt.figure(figsize=(10, 6))
plt.hist(delivery_times, bins=range(2, 10), edgecolor='black')
plt.title('Время доставки товаров')
plt.xlabel('Дни')
plt.ylabel('Количество заказов')
plt.show()

# Круговая диаграмма процента возвратов
total_items = sum(returns_percent) + len(returns_percent) * 100
wedges, texts, autotexts = plt.pie([sum(returns_percent), total_items - sum(returns_percent)],
                                   labels=['Возвраты', 'Не возвращено'],
                                   colors=['#ff9999', '#66b3ff'], 
                                   autopct=lambda p : '{:.1f}%'.format(p),
                                   startangle=90)
plt.axis('equal')
plt.title('Статус товаров после продаж')
plt.legend(wedges, [item.get_text() for item in texts], title="Статус",
           loc="center left", bbox_to_anchor=(1, 0, 0.5, 0))

plt.show()

# Ствол-лист диаграмма количества заказов
def stem_plot(data):
    plt.figure(figsize=(10, 6))
    plt.stem(range(len(data)), data, linefmt='k-', basefmt='w-')
    plt.title('Количество заказов в месяц')
    plt.xlabel('Месяц')
    plt.ylabel('Количество заказов')
    plt.show()

stem_plot(orders)
```

# Задание 5

Небольшой банк автоматизирует деятельность, связанную с кредитованием физических и юридических лиц (индивидуальных клиентов и организаций).
В настоящее время кандидат на получение кредита заполняет бумажную форму, прикладывает необходимые документы (финансовый отчет, перспективную оценку финансового состояния и др.) и отправляет в банк. Референт по кредитованию анализирует запрос на предмет возможных ошибок и подтверждает его достоверность.
Затем референт запрашивает отчет о кредитных операциях клиента в отделе кредитования. Копия отчета просматривается банковским служащим, а референт проверяет финансовое положение и доход клиента. Служащий также обращается к существующей системе управления счетами клиентов, чтобы получить необходимую информацию о состоянии счета и предыдущих кредитах клиента.
Вся информация комплектуется в кредитный запрос и направляется для оценки инспектору по кредитам. Если запрос утверждается, инспектор определяет наилучшие условия кредитования и уведомляет об этом клиента. Если клиент принимает условия, то кредит оформляется.
На обработку запроса обычно уходит минимум две недели (как для индивидуальных клиентов, так и для организаций).
Цель автоматизации сократить время обработки запроса до 48 часов для индивидуальных клиентов и 72 часов для организаций, сократить количество сотрудников, занятых в процессе обработки, и увеличить количество запросов, обрабатываемых в заданный период.



```python
import matplotlib.pyplot as plt
import numpy as np

# Подготовка данных
time_individual = np.linspace(0, 14, 15)  # Времени обработки для индивидуальных клиентов
time_organization = np.linspace(16, 30, 15)  # Времени обработки для организаций
num_requests_individual = np.random.randint(50, 150, 15)
num_requests_organization = np.random.randint(30, 80, 15)

# 1. Линейный график
plt.figure(figsize=(10, 6))
plt.plot(time_individual, num_requests_individual, label='Индивидуальные клиенты')
plt.plot(time_organization, num_requests_organization, label='Организации')
plt.xlabel('Время (дни)')
plt.ylabel('Количество запросов')
plt.title('Изменение количества обработанных запросов со временем')
plt.legend()
plt.grid(True)
plt.show()

# 2. Диаграмма разброса
plt.figure(figsize=(10, 6))
plt.scatter(time_individual, num_requests_individual, label='Индивидуальные клиенты')
plt.scatter(time_organization, num_requests_organization, label='Организации')
plt.xlabel('Время (дни)')
plt.ylabel('Количество запросов')
plt.title('Взаимосвязь времени обработки и количества запросов')
plt.legend()
plt.grid(True)
plt.show()

# 3. Столбчатая диаграмма
plt.figure(figsize=(10, 6))
plt.bar(['Индивидуальные клиенты', 'Организации'], [np.mean(num_requests_individual), np.mean(num_requests_organization)])
plt.xlabel('Тип клиента')
plt.ylabel('Среднее количество запросов')
plt.title('Сравнение среднего количества запросов')
plt.show()

# 4. Гистограмма
plt.figure(figsize=(10, 6))
plt.hist(time_individual, bins=10, edgecolor='black', label='Индивидуальные клиенты')
plt.hist(time_organization, bins=10, edgecolor='black', label='Организации')
plt.xlabel('Время обработки (дни)')
plt.ylabel('Частота')
plt.title('Распределение времени обработки запросов')
plt.legend()
plt.show()

# 5. Круговая диаграмма
labels = ['Индивидуальные клиенты', 'Организации']
sizes = [np.sum(num_requests_individual), np.sum(num_requests_organization)]
colors = ['#ff9999', '#66b3ff']
explode = [0, 0.1]  # Отступ для организации

plt.figure(figsize=(10, 8))
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Чтобы круговая диаграмма была круглой
plt.title('Доля различных типов клиентов')
plt.show()

# 6. Ствол-лист диаграмма
def stem_plot(data):
    plt.figure(figsize=(10, 6))
    plt.stem(range(len(data)), data)
    plt.xlabel('Порядковый номер')
    plt.ylabel('Обработанные запросы')
    plt.title('Ствол-лист диаграмма обработанных запросов')
    plt.grid(True)
    plt.show()

stem_plot(np.cumsum(num_requests_individual))
stem_plot(np.cumsum(num_requests_organization))
```

Компьютерная компания занимается продажей, ремонтом, сборкой, тестированием компьютерной техники. Также специалисты компании предоставляют услуги по разработке и монтажу локальных вычислительных сетей. Вся техника и комплектующие закупаются оптом у дилеров и хранятся на складе. Клиент, который хочет приобрести товар, оформляет заказ в торговом зале, а забирает технику со склада или оставляет заявку на ее доставку. Клиент, который хочет отремонтировать технику, приносит ее в сервисный отдел, откуда, по прошествии некоторого времени, забирает как отремонтированную или как технику, не подлежащую ремонту. По желанию клиента, специалисты компании могут выехать к клиенту для общей диагностики возникшей проблемы с техникой.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Подготовка данных
data = {
    'Sales': [1000, 1200, 1500, 1800, 2000],
    'Repair Services': [500, 550, 600, 650, 700],
    'Assembly Line Production': [800, 900, 1000, 1100, 1200],
    'Network Installation Services': [300, 350, 400, 450, 500],
    'Customer Diagnostics Visits': [50, 60, 70, 80, 90]
}

df = pd.DataFrame(data)


# Линейный график продаж
plt.figure(figsize=(10,6))
plt.plot(df.index, df['Sales'], marker='o')
plt.title('Динамика продаж компьютерной техники')
plt.xlabel('Месяцы')
plt.ylabel('Объем продаж')
plt.grid(True)
plt.show()

# Диаграмма разброса (scatter plot) для услуг ремонта
plt.figure(figsize=(10,6))
plt.scatter(df.index, df['Repair Services'], marker='s')
plt.title('Разброс услуг ремонта компьютерной техники')
plt.xlabel('Месяцы')
plt.ylabel('Объем услуг')
plt.grid(True)
plt.show()

# Столбчатая диаграмма производства на сборочной линии
plt.figure(figsize=(10,6))
plt.bar(df.index, df['Assembly Line Production'])
plt.title('Производство на сборочной линии')
plt.xlabel('Месяцы')
plt.ylabel('Объем производства')
plt.show()

# Гистограмма услуг по установке локальных сетей
plt.figure(figsize=(10,6))
plt.hist(df['Network Installation Services'], bins=5, edgecolor='black')
plt.title('Гистограмма услуг по установке локальных сетей')
plt.xlabel('Количество услуг')
plt.ylabel('Частота')
plt.show()

# Круговая диаграмма услуг диагностики
plt.figure(figsize=(10,6))
plt.pie(df['Customer Diagnostics Visits'], labels=df.index, autopct='%1.1f%%')
plt.title('Диагностика клиентов')
plt.axis('equal')  # Чтобы круговая диаграмма была кругом
plt.show()

# Столбчатая диаграмма с использованием seaborn
import seaborn as sns
plt.figure(figsize=(10,6))
sns.barplot(x=df.index, y=df['Repair Services'])
plt.title('Услуги ремонта компьютерной техники')
plt.xlabel('Месяцы')
plt.ylabel('Объем услуг')
plt.show()

# Линейный график для сравнения всех услуг
plt.figure(figsize=(15,8))
for service in df.columns[1:]:
    plt.plot(df.index, df[service], label=service)
plt.title('Сравнение различных услуг компании')
plt.xlabel('Месяцы')
plt.ylabel('Объем услуг')
plt.legend()
plt.grid(True)
plt.show()

# Диаграмма разброса для всех услуг
plt.figure(figsize=(15,8))
for service in df.columns[1:]:
    plt.scatter(df.index, df[service], marker='s', alpha=0.7)
plt.title('Разброс различных услуг компании')
plt.xlabel('Месяцы')
plt.ylabel('Объем услуг')
plt.grid(True)
plt.show()

# Гистограммы для всех услуг
plt.figure(figsize=(15,8))
for service in df.columns[1:]:
    plt.hist(df[service], bins=10, edgecolor='black', alpha=0.7)
plt.title('Гистограммы различных услуг компании')
plt.xlabel('Объем услуг')
plt.ylabel('Частота')
plt.legend(df.columns[1:])
plt.show()

# Круговая диаграмма для всех услуг
plt.figure(figsize=(15,8))
plt.pie(df.iloc[:, 1:].sum(), labels=df.columns[1:], autopct='%1.1f%%')
plt.axis('equal')
plt.title('Общий обзор услуг компании')
plt.show()
```
